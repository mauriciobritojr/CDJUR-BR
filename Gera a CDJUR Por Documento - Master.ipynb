```python
# Junta todas as anotações dos documentos numa lista
lista_ann = []
for doc in dict_cdjur_geral.keys():
    lista_ann.append(dict_cdjur_geral[doc])
```

```python
# Cria uma lista com as ocorrências de todas as anotações
lista_anotacoes = []
for ann in lista_ann:
    for dict in ann:
        lista_anotacoes.append(list(dict.keys())[0])
# Cria um conjunto das anotações para servir de chave para um dicionário
chaves = set(lista_anotacoes)
dict_aux = dict.fromkeys(list(set(chaves)), [])

for tag in dict_aux.keys():
    lista_aux = []
    quantidade = 0        
    for tag_ann in lista_anotacoes:
        if tag_ann == tag:
            quantidade += 1
    dict_aux[tag] = quantidade
```

```python
# Total de anotações por tag - Descritivo
dict_aux
```

```python
lst_grp = []
for tag in chaves:
    lst_grp.append(tag.split("_")[0]) # código original para criar os grupos de tags
lst_grp = list(set(lst_grp))
#cria dicionário para a totalização por grupo
dict_grp = dict.fromkeys(lst_grp, 0)
# Totaliza por grupo de tag
for tag in dict_aux.keys():
    dict_grp[tag.split("_")[0]] += dict_aux[tag]
dict_grp
```

```python
total = 0
for tag in dict_aux.keys():
    total += dict_aux[tag]
total
```

```python
# Cria uma lista de quantidade para o gráfico
lista_qtd = []
for tag in dict_aux.keys():
    lista_qtd.append(dict_aux[tag])
```

```python
tags = list(dict_aux.keys())
plt.figure(figsize=(12,16))
plt.barh(range(len(dict_aux)), lista_qtd, tick_label=tags)
plt.title("Ocorrências por Tag")


for indice,valor in enumerate(lista_qtd):
    plt.text(y=indice , x =valor+1 , s=f"{valor}")# , fontdict=dict(fontsize=20))

# plt.grid()
plt.show()
```

```python
# Gera arquivo .csv para o gráfico de pizza
df_cdjur = pd.DataFrame(list(zip(tags, lista_qtd)),
               columns =['tag', 'quantidade'])
# grava arquivo .csv
df_cdjur.to_csv('estatistica_cdjur_0911.csv', sep=',')
```
